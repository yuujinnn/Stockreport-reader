```python
from urllib.parse import quote
import requests
import json
import time
import os

def search_naver_news(keyword, max_count=50):
    headers = {
        "X-Naver-Client-Id": NAVER_CLIENT_ID,
        "X-Naver-Client-Secret": NAVER_CLIENT_SECRET
    }

    results = []
    start = 1
    while len(results) < max_count and start < 1000:
        url = f"https://openapi.naver.com/v1/search/news.json?query={quote(keyword)}&display=100&start={start}&sort=sim"
        response = requests.get(url, headers=headers)
        if response.status_code != 200:
            print(f"âŒ Error for '{keyword}' at start={start}: {response.status_code}")
            break

        items = response.json().get("items", [])
        for item in items:
            if "news.naver.com" not in item["link"]:
                continue  # ë„¤ì´ë²„ ë‰´ìŠ¤ë§Œ í•„í„°ë§
            results.append({
                "title": item["title"],
                "link": item["link"],
                "description": item["description"],
                "pubDate": item["pubDate"]
            })
            if len(results) >= max_count:
                break

        start += 100
        time.sleep(0.2)  # API rate limit ê³ ë ¤

    return results[:max_count]

def ensure_dir(path):
    directory = os.path.dirname(path)
    if not os.path.exists(directory):
        os.makedirs(directory)

# í‚¤ì›Œë“œë¥¼ ê³µë°±ìœ¼ë¡œ ì¡°í•©í•˜ì—¬ ë‹¨ì¼ ê²€ìƒ‰
def get_news(keywords, output_path):
    os.makedirs(os.path.dirname(output_path), exist_ok=True)

    combined_query = " ".join(keywords)
    print(f"ğŸ” Searching for combined keywords: {combined_query}")

    news_items = search_naver_news(combined_query)

    all_results = {"query": combined_query, "results": news_items}

    ensure_dir(output_path)
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(all_results, f, ensure_ascii=False, indent=2)

    print(f"ë„¤ì´ë²„ ë‰´ìŠ¤ ê¸°ì‚¬ {len(news_items)}ê°œ ì €ì¥ ì™„ë£Œ â†’ {output_path}")
```

```python
import json
import requests
from bs4 import BeautifulSoup

def crawling_news(articles, output_path):
    selected_news = [item for item in articles if item.get("selected") == 1]

    collected = []

    for i, article in enumerate(selected_news):
        link = article.get('link') or selected_news.get('ë§í¬')
        title = article.get('title') or selected_news.get('ì œëª©', '')

        if not link:
            print(f"[{i}] ë§í¬ ì—†ìŒ, ê±´ë„ˆëœ€")
            continue

        try:
            headers = {"User-Agent": "Mozilla/5.0"}
            res = requests.get(link, headers=headers, timeout=5)
            soup = BeautifulSoup(res.text, "html.parser")
            content = soup.select_one("#dic_area")
            content_text = content.get_text(separator="\n", strip=True) if content else "[ë³¸ë¬¸ ì—†ìŒ]"
        except Exception as e:
            print(f"[{i}] ë³¸ë¬¸ í¬ë¡¤ë§ ì‹¤íŒ¨: {e}")
            content_text = "[ë³¸ë¬¸ í¬ë¡¤ë§ ì‹¤íŒ¨]"

        collected.append({
            "ì œëª©": title,
            "ë§í¬": link,
            "ë³¸ë¬¸": content_text
        })
        print(f"[{i}] {title} - ë³¸ë¬¸ í¬ë¡¤ë§ ì™„ë£Œ")

    # ë£¨í”„ ë°–ì—ì„œ í•œ ë²ˆë§Œ ì €ì¥
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(collected, f, ensure_ascii=False, indent=2)

    print(f"\nğŸ“ í¬ë¡¤ë§ ì™„ë£Œ! ê²°ê³¼ ì €ì¥: {output_path}")
```
